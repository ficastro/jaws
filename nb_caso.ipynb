** Analisando os dados das olimpíadas de 2016 (Rio):**

```python
rio2016 = atletas[atletas['Year']==2016]
rio2016.head()
```

```python
# Quantas medalhas em Rio 2016?
rio2016['Medal'].value_counts()
```

```python
# Como é a distribuição de medalhas por esporte?
rio2016[rio2016['Medal'].notnull()].groupby(['NOC','Sport','Medal'])[['Event']].nunique()
```

```python
# Qual país ganhou cada medalha por tipo de jogo (evento)?

medalhas = atletas.groupby(['NOC','Team','Event','Medal']).nunique().index
medalhas = medalhas.to_frame()
medalhas.head()
```

* Argélia ganhou ouro e prata no 1500 metros no Rio2016.
* AHO: Antilhas Holandesas

```python
medalhas = medalhas.reset_index(drop=True)
medalhas.head()
```

```python
# Qual país ganhou cada medalha por tipo de jogo (evento)?
# Pivotagem para melhorar a visualização:

medalhas.pivot_table(values='NOC', index='Event', columns='Medal', aggfunc='first')
```

```python
# Qual o perfil dos vencedores (atletas que ganharam ouro)?

atletas_ouro = atletas[atletas['Medal']=='Gold'].copy()
atletas_ouro.head()
```
